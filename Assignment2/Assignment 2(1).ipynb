```python
import pandas as pd
import numpy as np
import numpy.random as rd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('airquality.txt',index_col=0)
df = pd.DataFrame(df)
df
```

```python
df.isnull().sum()
```

```python
mean=df.mean()
mean
```

```python
mean_Ozone=df['Ozone'].mean()
mean_Ozone
```

```python
df2=df['Ozone'].replace(np.nan,mean_Ozone)
df2
```

```python

df2=df.replace(np.nan,mean)
df2
```

```python

sns.lineplot(data=df2,x='Temp',y='Ozone')
```

```python
#sns.lineplot(data=df, x="Temp", y="Day", hue="Month")
sns.scatterplot(data=df2,x="Temp",y="Day",hue="Month")
```

```python
X=df2[['Temp']]
y=df2[['Ozone']]
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_predict=regressor.predict(X_test)
y_predict
```

```python
plt.scatter(X_train, y_train,color='g') 

plt.plot(X_test, y_predict,color='k') 
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,y_predict)
MSE
```

```python
from sklearn.metrics import mean_squared_error
RMSE=mean_squared_error(y_test,y_predict,squared=False)
RMSE
```

```python
import sklearn.metrics as met
r2score=met.r2_score(y_test,y_predict)
r2score
```
